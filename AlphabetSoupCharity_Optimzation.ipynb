```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("/content/drive/My Drive/charity_data.csv")
#application_df.head()
```

```python
application_df.drop(columns=['EIN','NAME'], inplace=True)
```

```python
optimization_1 = application_df.copy()
optimization_2 = application_df.copy()
```

#Optimization 1 - Original data with more Hidden Layers

```python
application_types_to_replace = ["T9", "T13", "T12", "T2", "T14","T25","T15","T29","T17"]

for app in application_types_to_replace:
    optimization_1['APPLICATION_TYPE'] = optimization_1['APPLICATION_TYPE'].replace(app,"Other")

classifications_to_replace = ["C7000","C1700","C4000","C5000","C1270","C2700","C2800","C7100","C1300","C1280","C1230","C1400","C7200","C2300","C1240","C8000","C7120","C1500","C1800","C6000","C1250","C8200","C1238","C1278","C1235","C1237","C7210","C2400","C1720","C4100","C1257","C1600","C1260","C2710","C0","C3200","C1234","C1246","C1267","C1256","C2190","C4200","C2600","C5200","C1370","C1248","C6100","C1820","C1900","C1236","C3700","C2570","C1580","C1245","C2500","C1570","C1283","C2380","C1732","C1728","C2170","C4120","C8210","C2561","C4500","C2150"]

for cls in classifications_to_replace:
    optimization_1['CLASSIFICATION'] = optimization_1['CLASSIFICATION'].replace(cls,"Other")
```

```python
optimization_1_dummies = pd.get_dummies(optimization_1)
```

```python
X_optimization1 = optimization_1_dummies.drop(columns=['IS_SUCCESSFUL'])
y_optimization1 = optimization_1_dummies['IS_SUCCESSFUL']

X_train1, X_test1, y_train1, y_test1 = train_test_split(X_optimization1, y_optimization1, random_state=35)
```

```python
scaler1 = StandardScaler()

X_scaler1 = scaler1.fit(X_train1)

X_train_scaled1 = X_scaler1.transform(X_train1)
X_test_scaled1 = X_scaler1.transform(X_test1)
```

```python
nn1 = tf.keras.models.Sequential()

nn1.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=43))

nn1.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=22))
nn1.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=11))
nn1.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=5))
nn1.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=2))


nn1.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Check the structure of the model
nn1.summary()
```

```python
nn1.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
fit_model1 = nn1.fit(X_train_scaled1, y_train1, epochs=100)
```

```python
model_loss, model_accuracy = nn1.evaluate(X_test_scaled1,y_test1,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
from pandas.io.pytables import HDFStore

nn1.save('optimizedModel_1.h5')
```

###Observations
Training model accuracy maxes out around .7393 at epoch 75/100. More layers does not solve the issue
